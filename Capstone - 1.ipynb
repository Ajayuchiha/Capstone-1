```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Project\New_folder_project\orders.csv")
```

```python
#dropping null
```

```python
print ((df.isnull().sum()))
```

```python
(df.dropna(inplace = True))
```

```python
print ((df.isnull().sum()))
```

```python
#columns renaming
```

```python
df = df.rename(columns = {"Order Id" : "order_id", "Order Date" : "order_date", "Ship Mode" : "ship_mode", 
                              "Segment" : "segment", "Country" : "country", "City" : "city", "State" : "state", 
                              "Postal Code" : "postal_code", "Region" : "region", "Category" : "category", 
                              "Sub Category" : "sub_category", "Product Id" : "product_id",  "cost price" : "cost_price", 
                              "List Price" : "list_price", "Quantity" : "quantity", "Discount Percent" : "discount_percentage"})
```

```python
#To find discount 
df['discount'] = df['list_price'] * df['discount_percentage'] / 100
```

```python
#To find sale price
df['sale_price'] = df['list_price'] - df['discount']
```

```python
#To find profit
df['profit'] = df['sale_price'] - df['cost_price']
```
